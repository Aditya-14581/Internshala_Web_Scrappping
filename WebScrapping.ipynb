```python
# for i in soup.find_all('p'):
#   print(i.text.strip())

# print(soup.find_all('div',class_='individual_internship')[0].text.strip())
companies = soup.find_all('div',class_='internship_meta')
base_url = "https://internshala.com"
# p=soup.find_all('a',class_="view_detail_button_outline")
# print(base_url + p.get('href'))
apply_link = []
for link in soup.find_all('a',class_="view_detail_button_outline"):
  apply_link.append(base_url + link.get('href'))

len(apply_link)
```

```python
def get_element_text(element, default=""):
    """
    Checks if an element is valid and returns its text content if it exists, or a default value otherwise.

    Args:
        element: BeautifulSoup element to extract text from.
        default: Default value to return if element is invalid or does not contain text. Default is "Not Available".

    Returns:
        Text content of the element if it exists, otherwise the default value.
    """
    if element:
        return element.text.strip()
    else:
        return default
```

```python
name = []
role = []
wfh = []
duration = []
stipend = []


for i in companies:
    name.append(i.find('p').text.strip())
    # type of roles
    role_element = i.find('h3', class_='profile')
    role.append(get_element_text(role_element))

    # work form home or not
    wfh_element = i.find('a')
    wfh.append(get_element_text(wfh_element))
    # duration
    duration.append(i.find_all('div', class_='item_body')[1].text.strip())
    stipend.append(i.find_all('div', class_='item_body')[2].text.strip())

# companies
# role
# wfh
# duration
# view_details
# len(name)
# stipend

df=pd.DataFrame({'Company':name,
                 'Role':role,
                 'Location':wfh,
                 'Duration':duration,
                 'Stipend' : stipend,
                 'Apply Link':apply_link,
   })

# print(len(name))
# print(len(role))
# print(len(wfh))
# print(len(duration))
# print(len(stipend))
# print(len(apply_link))
df.shape
```

```python
final=pd.DataFrame()
for j in range(1,5):
  # webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
  webpage=requests.get('https://internshala.com/internships/computer-science-internship/page-{}/'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  companies = soup.find_all('div',class_='internship_meta')
  base_url = "https://internshala.com"

  apply_link = []
  for link in soup.find_all('a',class_="view_detail_button_outline"):
    apply_link.append(base_url + link.get('href'))

  name = []
  role = []
  wfh = []
  duration = []
  stipend = []


  for i in companies:
      name.append(i.find('p').text.strip())
      # type of roles
      role_element = i.find('h3', class_='profile')
      role.append(get_element_text(role_element))

      # work form home or not
      wfh_element = i.find('a')
      wfh.append(get_element_text(wfh_element))
      # duration
      duration.append(i.find_all('div', class_='item_body')[1].text.strip())
      stipend.append(i.find_all('div', class_='item_body')[2].text.strip())



  df=pd.DataFrame({'Company':name,
                  'Role':role,
                  'Location':wfh,
                  'Duration':duration,
                  'Stipend' : stipend,
                  'Apply Link':apply_link,
    })

  final=final.append(df,ignore_index=True)
```
